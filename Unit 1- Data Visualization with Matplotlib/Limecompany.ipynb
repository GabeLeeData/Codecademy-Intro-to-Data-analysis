Unit 1 Data Visualization with Matplotlib Project 2: Sublime Limes' Line Graphs

This project would involve using matplotlib to create different graphs that involve some data regarding a company that sells Limes.

```python
%matplotlib inline
```

```python

from matplotlib import pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]

# numbers of limes of different species sold each month
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]
persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]
blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]



# create your figure here
plt.figure(figsize=(12,8))

#Plot 1
ax1 = plt.subplot(1,2,1)
x_values = range(len(months))
plt.plot(x_values, visits_per_month, marker='o')
ax1.set_xticklabels(months)
plt.xlabel('Months')
plt.ylabel('Visits Per Month')
plt.title('Visits of Websites Per Month')
ax1.set_xticks(x_values)


#Plot 2
ax2 = plt.subplot(1,2,2)
plt.plot(x_values, key_limes_per_month, color='green', marker='o')
plt.plot(x_values, persian_limes_per_month, color='yellow', marker='o')
plt.plot(x_values, blood_limes_per_month, color='red', marker='o')
ax2.set_xticklabels(months)
plt.xlabel('Months')
plt.ylabel('Limes Sold')
ax2.set_xticks(x_values)
legend_labels = 'Key Limes', 'Persian Limes', 'Blood Limes'
plt.title("Lime Sales Per Month")
plt.legend(legend_labels)



plt.subplots_adjust(wspace = 0.35, bottom = 0.2)

#Save Figure
#plt.savefig("sublime_limes_line_graphs.png")

plt.show()
```
